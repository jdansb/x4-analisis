Para realizar essa conversão para uma unidade observável no mapa, marquei uma distância de 14 quadrados e medi o tempo necessário para o camponês percorrê-la. Com isso, temos então que o tempo total de produção de 15 unidades de comida é 45s para um fazendeiro. Sendo assim, cada unidade de comida tem 3s de trabalho contido, ou de outra orma, a cada 3 segundos temos uma unidade de comida

Se temos então $F$ fazendeiros, temos uma produção bruta total por segundo de $F/3$, com isso a cada três segundos  é produzido $3 \cdot F/3=F$ unidades de comida. Mas  cada fazendeiro consome uma unidade de comida a cada 10 segundos, ou seja, a cada segundo $F/10$ unidades de comida são consumidas pelos mesmos fazendeiros que a produzem. Então por segundo a produção líquida dos fazendeiros é:

$$P_L=F/3-F/10=F(1/3-1/10)=F(10-3)/30$$

Ou apenas:

$$P_L=F\frac{7}{30}$$

Uma forma de ver isso é que para cada trabalhador, ele produz $7/30$ unidade de comida por segundo, ou $7$ unidades pra cada 30 segundos, ou ainda 14 unidades de comida por minuto.


Ignorando questões específicas de estratégia militar como necessidade de diversificar as unidades militares devido a vantagens e desvantagens entre si, vamos considerar apenas que uma maior produção econômica é capaz de suportar um maior exército. Vamos considerar então apenas a produção militar de um tipo de unidade: a carruagem. Cada carruagem consome 4 unidades de comida a cada 10 segundos. Então ao total as carruagens vão consumir $4C/10$ unidades de comida por segundo. A produção líquida total de comida na minha sociedade é então:

$$P=F\frac{7}{30}-C\frac{4}{10}$$.

Considerando uma situação que não precisamos de excedente, então $P=0$, de forma que podemos ter:

$$0=F\frac{7}{30}-C\frac{4}{10}$$.

Ou ainda:

$$C=F\frac{7}{12}\approx0.5F$$.


Ou seja, precisamos aproximadamente de 2 fazendas pra sustentar uma caruagem. Levando em conta que cada carruagem consome 3 espaços de população, então precisamos de 5 espaço populacional. Esta é uma espécie de unidade mínima sustentável de nossa sociedade: o conjunto de 2 fazendas, 5 espaço populacionais, 2 fazendeiros e uma carruagem. A partir disso podemos expandir a sociedade de forma sustentável.

Se temos 10 fazendas, esperamos então sustentar aproximadamente 5 carruagens, ocupando $5\cdot 3+10=25$ espaço da população. A produção líquida de comida por segundo, nesse caso, será:

$$P=10\frac{7}{30}-5\frac{4}{10}=0.33$$

Após cinco minutos esperaríamos então $5 \cdot 60 \cdot 0.33= 100$ novas unidades de comida estocadas.

Porém meu teste, obtivemos $440$. Isto equivale a $\frac{440}{300}=1.46$ unidade de comida por segundo. Esta quantidade adicional de comida deve-se ao fato que em média as fazendas estão mais próximas do armazém do que nossa suposição inicial (15 quadrados). A produção bruta real foi então a soma da comida:

- Consumida pelas 5 carruagens $5 \cdot 300 \cdot 4/10 =600$ no mesmo período total.
- Excedente de $440$

Então a produção bruta total é de $600+300+440=1340$. Como são 10 fazendeiros, então cada fazendeiro produziu em média $134$ unidades de comida.

Isso significa que foi necessário $300/134\approx 2.24$ segundos pra produzir cada unidade de comida, ou colocando de outra forma, cada unidade de comida tem $2.24$ segundos de trabalho contido. Dessa forma 15 unidades de comida é produzida a cada $\approx 33.6$ segundos. Utilizo sempre 15 unidades para análise pois é a capacidade máxima de transporte de cada camponês, isto é, após ele colher 15 unidades de comida ele se dirige ao armazém mais próximo para depositá-la e retornar a sua fazenda para reiniciar a produção.

Como temos os mesmos 30 segundos pra cultivo, que já tinha sido uma grandeza observada,  então o tempo de deslocamento real foi $3.6$ segundos no total. Então a distância entre o armazém e a fazenda foi percorrida em média em $1.8$ segundos por cada fazendeiro. Como ele percorre 2 quadrados por segundo, então a distância real, em média, das fazendas para os armazéns é de $2*1.8=3.6$ quadrados.  Significativmente menor que nossa estimativa anterior.

Vamos corrigir nosso estudo inicial. Se agora 15 unidades de comida é produzido a cada 33.6 segundos então a produção lúquida dos fazendeiros é:

$$P_L=F \cdot \frac{15}{33.6}-\frac{F}{10}=F( \frac{15}{33.6}-\frac{1}{10})\approx 0.35 F$$

E a produção líquida total da sociedade considerando fazendeiros e carruagens é de:
$$P=0.35F-C\frac{4}{10}$$.

Com 10 fazendeiros e 5 carruagens temos:
$$P=10 \cdot 0.35-5\frac{4}{10}\approx 1,5$$

Recuperamos aproximadamente o resultado medido empiricamente de um excedente aproximado de 1.5 unidade de comida por segundo, com esta específica disposição de fazendas e armazéns, além da distribuiçao populacional.

Uma lição que fica é que as vezes ao invés de tentarmos 'adivinhar' uma estimativa (como a distância entre armazéns e fazenda), e planejar tudo com antecedencia, é melhor construirmos uma situação real e coletarmos dados sobre esta situação para decidirmos sobre o futuro.

A análise construída até aqui, foi feita digamos, respondendo uma demanda artificial, de puramente curiosidade. Agora que comecei a campanha, novas demandas devem surgir.

camponeses, incluindo o próprio fazendeiro.
