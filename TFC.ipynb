Hoje eu quero fazer uma análise de [*The Fertile Crescent*](https://store.steampowered.com/app/1674820/TFC_The_Fertile_Crescent/). Minha meta é usar o jogo para aplicar algumas ferramentas que já estudei, ao mesmo tempo em que exploro novos recursos analíticos. Um dos principais desafios aqui é a baixa estabilidade do sistema, sobretudo porque não temos controle sobre alguns recursos essenciais (como madeira), que se esgotam relativamente rápido. Isso dificulta encontrar qualquer janela temporal em que seja possível aproximar um estado de estabilidade.

Por isso, optei por focar em uma economia baseada na produção de comida, já que esse recurso pode ser tratado como efetivamente não esgotável dentro do jogo. Eu construí um mercado onde podemos realizar as seguintes trocas:

|    | Comprar |    | Vender  |
|----|---------|----|---------|
| 10 | Comida  | 25 | Madeira |
| 10 | Comida  | 15 | Metal   |
| 10 | Madeira | 10 | Comida  |
| 10 | Metal   | 20 | Comida  |

De cara podemos notar que as trocas não são “justas”: entregamos 15 de metal para obter 10 de comida. Ou seja, cada unidade de comida custa 1,5 metal se quisermos comprar comida.

Por outro lado, entregamos 20 de comida para obter 10 de metal. Assim, se quisermos comprar metal, cada unidade de comida vale 0,5 metal nesse sentido de troca.

Escolhendo a civilização egípcia e considerando todas as pesquisas realizadas, observamos que um camponês leva aproximadamente 30 segundos para coletar 15 unidades de comida em uma fazenda. Entretanto, também é necessário considerar o tempo de deslocamento entre a fazenda e o depósito mais próximo. Vamos estimar esse tempo em cerca de 15 segundos.

Considerando que a velocidade de cada unidade é igual a 9 precisamos traduzir isso para alguma velocidade. Para converter essa medida para uma unidade observável no mapa, marquei uma distância de 14 quadrados e medi o tempo necessário para o camponês percorrê-la. O resultado foi de 7 segundos, o que corresponde a uma velocidade de 2 quadrados por segundo. Se 2 quadrados por segundo equivalem a velocidade 9, consequentemente, os 15 segundos extra de deslocamento correspondem a um percurso total de 30 quadrados. Como esse valor inclui ida e volta, a distância entre a fazenda e o depósito é de aproximadamente 15 quadrados.

Com isso, temos então que a produção é de 15 unidades de comida a cada 45s. Sendo assim, cada unidade de comida tem 3s de trabalho contido. Ou a cada 3segundos temos uma unidade de trabalho contido.

Se temos então $F$ fazendeiros, temos uma produção bruta total por segundo de $F/3$. A cada três segundos $3 \cdot F/3=F$ unidades de comida é produzida. Mas  cada unidade consome uma unidade de comida a cada 10 segundos, ou seja, a cada segundo $F/10$ unidades de comida são consumidas pelos fazendos. Então por segundo a produção líquida dos fazendeiros é:

$$P_L=F/3-F/10=F(1/3-1/10)=F(10-3)/30$$

Ou apenas:

$$P_L=F\frac{7}{30}$$

Ou seja , para cada trabalhador, ele produz $7/30$ unidade de comida por segundo, ou $7$ unidades pra cada 30 segundos ou 14 por minuto.


Vamos ignorar questões de estratégia militar e a necessidade de diversificar as unidades militares, também vamos ignorar a disponibilidade de terreno para cultivo. Vamos apenas considerar a produção militar de uma unidade: carruagem. Cada carruagem consome 4 unidades de comida a cada 10 segundos. Então ao total as carruagens vão consumir $4C/10$ unidades de comida por segundo. A produção líquida total de comida é então:

$$P=F\frac{7}{30}-C\frac{4}{10}$$.

Considerando uma situação que não precisamos de excedente, então $P=0$, de forma que podemos ter:

$$0=F\frac{7}{30}-C\frac{4}{10}$$.

Ou:

$$C=F\frac{7}{12}\approx0.5F$$.


Ou seja, precisamos de 2 fazendas pra sustentar uma caruagem aproximadamente. E levando em conta que cada carruagem consome 3 espaços de população, então precisamos de 5 espaço populacional. Se temos 10 fazendas, esperamos então sustentar aproximadamente 5 carruagens, ocupando $5\cdot 3+10=25$ espaço da população. A produção líquida de comida por segundo será:

$$P=10\frac{7}{30}-5\frac{4}{10}=0.33$$

Lembrando que isso é por segundo. Após cinco minutos esperaríamos então $5 \cdot 60 \cdot 0.33= 100$ novas unidades de comida. No meu teste, obtivemos $440$. Isto equivale a $\frac{440}{300}=1.46$ unidade de comida por segundo. Esta unidade de adicional de comida deve-se ao fato que em média as fazendas estão mais próximas do armazem do que nossa suposição inicial. A produção bruta foi então o consumido pelos 10 fazendeiros em 300 segundos $10 \cdot 300 \cdot 1/10=300$, o consumido pelas 5 carruagens $5 \cdot 300 \cdot 4/10 =600$, ou seja a comida consumida no total $600+300=900$, mais o excedente: $900+440=1340$. Como são 10 fazendeiros, então cada fazendeiro produziu em média $134$ unidades de comida.

Isso significa que $300/134\approx 2.24$ segundos pra produzir cada unidade de comida. Ou que 15 unidades de comida é produzida a cada $\approx 33.6$ segundos. Como temos fixo 30 segundos pra produção, então o tempo de deslocamento foi $3.6$ segundos, ou seja, a distância do armazém ta a $1.8$ segundos. Como ele percorre 2 quadrados por segundo, então a distância das fazendas para o armazém é em me´dia $2*1.8=3.6$ quadros.

Vmos retomar o estudo inicial. Se agora 15 unidades de comida é produzido a cada 33.6 segundos então:

$$P_L=F \cdot \frac{15}{33.6}-\frac{F}{10}=F( \frac{15}{33.6}-\frac{1}{10})\approx 0.35 F$$

Atualizando tudo então:
$$P=0.35F-C\frac{4}{10}$$.

Retomando então por segundo:
$$P=10 \cdot 0.35-5\frac{4}{10}\approx 1,5$$.

Evidentemente, só recuperamos aproximadamente o resultado medido empiricamente. Temos um excedente aproximado então de 1.5 unidade de comida por segundo, com esta disposição das fazendas e armazéns, e 5 carrguagens.

Testando então pra 10 minutos, esperamos $10\cdot 60\cdot 1.5 = 900$ de excedente. Empiricamente medimos: 881. O que torna razoável nossa estimativa que nossas fazendas estão em média, a 3.6 quados, considerando que cada fazendeiro percorre 2 quadros por segundo.
